```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Asignación de variables

```python
h=0.01
T=np.arange(5,15,h)
theta=np.zeros(len(T))
v=np.zeros(len(T))
v[0]=100
theta[0]=1000
w=2*3.14
```

# Formulas dadas en clase 

```python
for i in range(800):
    theta[i+1]=theta[i]+(h*v[i])
    v[i+1]=v[i]+(h*(-w**2*theta[i]))
```

# Gráfica 1

```python
plt.plot(T,v)#grafica del tiempo contra la velocidad
plt.xlabel('tiempo')
plt.ylabel('velocidad')
```

# Gráfica 2

```python
plt.plot(T,theta, color= 'g') #grafica del tiempo contra theta
plt.xlabel('tiempo')
plt.ylabel('theta')
```

# Gráfica 3

```python
plt.plot(v,theta, color='y') #velocidad contra el angulo
plt.xlabel('velocidad')
plt.ylabel('theta')
```
